```python
# Ingreso de sueldos de Juan durante 3 meses
sueldos = []
for i in range(3):
    sueldo = float(input(f"Ingrese el sueldo de Juan en el mes {i+1}: "))
    sueldos.append(sueldo)

# Cálculo del sueldo promedio de Juan
sueldo_promedio = sum(sueldos) / len(sueldos)

# Determinación del nivel de sueldo
nivel_sueldo = ""
if sueldo_promedio < 100000:
    nivel_sueldo = "bajo"
elif sueldo_promedio >= 100000 and sueldo_promedio <= 200000:
    nivel_sueldo = "normal"
else:
    nivel_sueldo = "mejor de lo normal"

# Mostrar resultados
print(f"El sueldo promedio de Juan es: {sueldo_promedio:.2f}")
print(f"El nivel de sueldo de Juan es: {nivel_sueldo}")
```
